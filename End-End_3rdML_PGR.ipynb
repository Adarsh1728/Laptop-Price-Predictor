```python
def fetch_processer(text):
    text = text.lower()
    if 'intel' in text:
        return 'Intel'
    elif 'amd' in text:
        return 'AMD'
    elif 'apple' in text:
        return 'Apple'
    else:
        return 'Other'
```

```python

df['cpu_brand'] = df['cpu_name'].apply(fetch_processer)
df.head(15)
```

```python
def os_cat(inp):
    if inp==' Windows 10' or inp=='Windows 7' or inp=='Windows 10 S':
        return " windos"
    else :
        if inp=="macOS" or inp=="macOS 10 S":
            return "mac"
        else:
            return "other No OS/Chrome OS/Android"
```

```python
df['Os']=df['OpSys'].apply(os_cat)
df
```

```python
x=df.drop(columns=['Price'])
y=np.log(df['Price'])
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
# Column transformer
# Identify categorical columns
categorical_cols = x_train.select_dtypes(include='object').columns.tolist()
print("Categorical columns:", categorical_cols)

# Column transformer with handle_unknown
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_cols)
], remainder='passthrough')
# Model
step2 =RandomForestRegressor(
    n_estimators=100,
    max_depth=50,
    max_samples=0.5,
    max_features=0.75,
    random_state=3,
    bootstrap=True
)

# Pipeline
Pipe= Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Fit
print(len(x_train.columns))
print(x_train.columns)

Pipe.fit(x_train,y_train)
y_pred = Pipe.predict(x_test)

# Metrics
print("R2 Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```
